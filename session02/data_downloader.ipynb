```python
URL = "https://raw.githubusercontent.com/hovhannisyan91/tableau-analytics-portfolio/refs/heads/main/session02/data"

import pandas as pd
tables = [
    "customers.csv",
    "customer_addresses.csv",
    "orders.csv",
    "order_details.csv",
    "products.csv",
    "returns.csv",
    "salespeople.csv"
]

for table in tables:
    url = f"{URL}/{table}"
    df = pd.read_csv(url)
    df.to_csv(f"data/{table}", index=False)
```
